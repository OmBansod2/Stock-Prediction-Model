```python
# Define a function to create a CSV file if it doesn't exist
def create_or_append_to_csv(data, L2):
    try:
        # Append to CSV if file exists
        data.to_csv(L2, mode='a', index=False, header=False)
    except FileNotFoundError:
        # Create a new CSV file if it doesn't exist
        data.to_csv(L2, index=False)
```

```python
def createframe(msg):
    df = pd.DataFrame([msg])
    df = df.loc[:,['s','E','p']]
    df.columns = ['symbol','Time','Price']
    df.Price = df.Price.astype(float)
    df.Time = pd.to_datetime(df.Time, unit='ms')
    return df
```

```python
# Define global variable to keep track of status
#global stat
stat = 'n'  # n: neutral, b: buy, s: sell
global pur_val # Initialize purchase value
pur_val = 0 
global account_balance 
account_balance = 5
aval_stock = 0
a1 = 0
s_price = 0
s_amt = 0
f = 0
g = 0
y = 0
oy = 0
ot = None

# Define your buy and sell functions here
def buy(xt,money):
    
    long(client, sym, quantity)
    global account_balance
    global aval_stock
    global a1
    global s_price
    global s_amt
    global stat
    global pur_val
    print("Buying...")
    price=xt
   
    if stat=='n': # Buying directly
        # Calculate the total cost of the purchase
      
        pur_val=xt
        amount=money/price
        total_cost = amount * price
        print("cost=", total_cost)
        # Check if the account balance is sufficient for the purchase
        if total_cost <= account_balance:
            # Deduct the total cost from the account balance
            # order = client.create_order(symbol='BTCUSDT',side='BUY',type='MARKET',quality=amount)
            # print(order)
            account_balance -= total_cost
            aval_stock = amount
            print("avail_stock=",aval_stock)
            print(f"Bought {amount} units at ${price} each. Total cost: ${total_cost}. Account balance: ${account_balance}")
            # return True
        else:
            print("Insufficient funds to buy.")
            # return False

    elif stat=='s': # Buying after selling
        # Calculate the total revenue from the sale
        # order = client.create_order(symbol='BTCUSDT',side='BUY',type='MARKET',quality=s_amt)
        # print(order)
        b_price = s_amt * price
        Tot_profit = s_price - b_price
        print("PROFIT=",Tot_profit)
        account_balance = account_balance + Tot_profit + s_price

        # print(f"Bought {amount} units at ${price} each. Total cost: ${b_price}. Account balance: ${account_balance}")
        print(f"Account balance: ${account_balance}")
        #return True

    # Update status based on current status
    #global stat
    if stat == 'n':
        stat = 'b'
    elif stat == 's':
        stat = 'n'
    else:
        exit()  # Exit if status is unexpected

def sell(xt,money):
    print("Selling...")
    global account_balance
    global aval_stock
    global a1
    global s_price
    global s_amt
    global stat
    global pur_val
    price = xt
    
    if stat=='n': # Selling directly(without buying)
        
        pur_val=xt
        amount=money/price
        total_cost = price * amount
        # order = client.create_order(symbol='BTCUSDT',side='SELL',type='MARKET',quality=amount)
        # print(order)
        print("cost=", total_cost)
        account_balance = account_balance - total_cost
        print(f"Sold {amount} units at ${price} each. Account balance: ${account_balance}")
        a1=price 
        s_price = price * amount
        s_amt = amount
    
    elif stat=='b': # Selling after buying
        # Calculate the total cost of the purchase
        total_revenue = aval_stock * price
        # order = client.create_order(symbol='BTCUSDT',side='SELL',type='MARKET',quality=aval_stock)
        # print(order)
        print("REV=",total_revenue)
        account_balance = account_balance+total_revenue
        # print(f"Sold {amount} units at ${price} each. Total revenue: ${total_revenue}. Account balance: ${account_balance}")
        print(f"Account balance: ${account_balance}")
        # return True
        
    # Update status based on current status
    #global stat
    if stat == 'n':
        stat = 's'
    elif stat == 'b':
        stat = 'n'
    else:
        exit()  # Exit if status is unexpected


# Define trading_bot function here
def trading_bot(x_value, initial_z, initial_oz, TIME):
    global stat, pur_val,account_balance  # Mark global variables
    
    # Initialize variables
    z = initial_z
    oz = initial_oz
    global f 
    global g 
    global y 
    global oy 
    global ot 
    money = account_balance
    print(money)
    print("f=",f)
    print("g=",g)
    print("stat=",stat)
    # Iterate over the values of the variable over time
    
    xt = x_value[1]
    xt_1 = x_value[0]

    # Check if current value is greater than previous value
    if xt > xt_1:
        z = z + (xt - xt_1)

        # Check if z exceeds oz
        if z > oz:
            f += 1
            g = 0
            # Trigger buy action if f reaches 10 and status is not 'b'
            if f == 10 and stat != 'b':
                buy(xt,money)
                f = 0
                ot = TIME
                oz = z
                pur_val = xt
        
        elif oz<z:
            g -= 1
    elif xt < xt_1:
            # Adjust z based on the difference between current and previous values
            z = z - (xt_1 - xt)
            y = (-1) * z

            if oz > z:
                g += 1
                f = 0

                if g == 10 and stat != 's':
                    sell(xt,money)
                    g = 0
                    ot = TIME
                    oy = y     
            elif z>oz:
                f -= 1

    # Update ot and oz if f reaches 0
    if f == 0:
        ot = TIME
        oz = z

    # Trigger sell action if f falls below -5 or xt is less than or equal to pur_val
    if (g >= 5 or xt <= (0.4 * pur_val)) and stat == 'b':
        sell(xt,money)  # Call sell function
        f = 0  # Reset counter
    
    # Trigger buy action if g reaches 5 and status is not 'b'
    if (f >= 5 or xt >= (1.4 * pur_val)) and stat == 's':
        buy(xt,money)  # Call buy function
        g = 0  # Reset counter
        
    return None

```
